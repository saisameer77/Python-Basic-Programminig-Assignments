```python
# Question 1

def div_by_7(n):
    

    for i in range(n):
        
    
        if  i%7 == 0:
            yield i
            
n = int(input())

for i in div_by_7(n):
    print(i)
    
    
```

```python
#Question 3

class Person():
    
    def __init__(self):
        
        print("class person is created")
        
class Male(Person):
    
    def __init__(self,name):
        
        self.name = name
        
        Person.__init__(self)
        
        
    def getGender(self):
        
        print(self.name, "is Male")
        
        

class Female(Person):
    
    def __init__(self,name):
        
        self.name = name
        
        Person.__init__(self)
        
        
    def getGender(self):
        
        print(self.name, "is Female")
        

       
        
m1 = Male(input("Enter the name of Male :"))

f1 = Female(input("Enter the name of Female :"))

m1.getGender()

f1.getGender()

```

```python
#Question 6

def Binaryfunc(List):


    binary = int(input("enter a number: "))


    s = int((len(List))/2)

    while True:
        
        if List[s] == binary :
            
            return s
        
        elif binary > List[s] :
            
            s = s + int((len(List) - s)/2) 
            
            
        elif binary < List[s]:
            
            s = int(s/2)

            
    
            
            
            
            
            


List = [56,34,78,43,86,87,90,200]

List.sort()

print("Sorted List is :", List)

ind = Binaryfunc(List)

print("Number found at index :", ind)
```
