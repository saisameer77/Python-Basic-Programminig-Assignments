```python
# Python program to check if a given string is binary string or not

String = input("Enter a string : ")

c = 0

for i in range(len(String)):
    
    if String[i] == "1":
        pass
    elif String[i] == "0":
        pass
    else:
        c = 1
        


if c == 0:
    print("String is a binary string")
    
else:
    print("String is not a binary string")
```

```python
# Python program to find duplicate characters from a string


st1 = input()


n = []



for i in range(len(st1)):
    
    m = 0
    
    for j in range(len(st1)):
        
        if st1[i] == st1[j]:
            
            m = m+1
            
        else :
            
            pass
        
        
    if m >= 2:
        
        
        n.append(st1[i])
        
print(set(n))
        
    
        
        
        
        
        
```
